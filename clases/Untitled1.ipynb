grafica ciudades



```
import matplotlib.pyplot as plt

ciudades =['Medellin','Bogota','Cali','Cartagena' ]
poblacion=[2508500, 7181469,2420100,914552]

plt.bar(ciudades,poblacion)
plt.title('Ciudades vs población')
plt.xlabel('Ciudades colombia')
plt.ylabel('Población')
plt.savefig('GraficoCiudades.png')
plt.show()
```

grafico tangente

```
import matplotlib.pyplot as plt
y = [0,0.26,0.57,1,1.73,3.73,0,-3.73,-1.73,-1,-0.57,-0.26,0,0.26,0.57,1,1.73,3.73,0,-3.73,-1.73,-1,-0.57,-0.26,0]
x = [0,15,30,45,60,75,90,105,120,135,150,165,180,195,210,225,240,255,270,285,300,315,330,345,360]


plt.plot(x,y,'g')
plt.title('Gráfico tangente')
plt.xlabel ('para valores de x avanzando en 15°')
plt.ylabel('valores de y')
plt.legend('f(x)')
plt.savefig('graficotangente.png')
plt.show()
```

alimentos

```
import matplotlib.pyplot as plt
preguntaIngresarAlimento = 'Ingrese el nombre del alimento,Ingresa de primero tu alimento favorito310 : '
preguntaIngresarAlimentoOcupacion = 'Ingrese %ocupación del alimento :  '
alimentos = []
ocupaciones = []
explodeList = [0.12,0,0,0,0,0]
cantidad = 6

for i in range(cantidad):
  alimento = input(preguntaIngresarAlimento)
  alimentos.append(alimento)
  ocupacion = int (input(preguntaIngresarAlimentoOcupacion))
  ocupaciones.append(ocupacion)

plt.title('Alimentos mas comunes vs ocupacion')
plt.pie(ocupaciones, labels= alimentos, shadow=True,explode= explodeList, startangle=45)
plt.savefig ('cajaalimentos.png')
plt.show()
```
